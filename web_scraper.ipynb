```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/List_of_inventors"

# Get the webpage content
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
# Find all <ul> elements
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Find the specific div
content_div = soup.find("div", class_="mw-content-ltr mw-parser-output")

if content_div:
    # Find all <ul> elements inside this div
    ul_elements = content_div.find_all("ul")

    for ul in ul_elements:
        li_elements = ul.find_all("li")  # Get all <li> inside the current <ul>

        for li in li_elements:
            print(li.text.strip())  # Print the text of each <li>
else:
    print("Content div not found.")

```
